```python
import numpy as np
import pandas as pd
import sklearn

from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_csv("C://Users/whs38//OneDrive//바탕 화면//archive//retail_dataset.csv",sep=",")
```

```python
items=set()

for col in df:
    items.update(df[col].unique())
```

```python
itemset=set(items)
encoding=[]
for index,row in df.iterrows():
    rowset=set(row)
    labels={}
    dismatching=list(itemset-rowset)
    matching=list(itemset.intersection(rowset))
    for i in dismatching:
        labels[i]=0
    for j in matching:
        labels[j]=1
    encoding.append(labels)
encoding[0]
result=pd.DataFrame(encoding)
result
```

```python
result=result.drop(result.columns[0],axis=1)
result
```

```python
fren_items=apriori(result,min_support=0.2,use_colnames=True)
```

```python
fren_items
```

```python
rules=association_rules(fren_items,metric="confidence",min_threshold=0.6)
rules
```
