```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

meat_consumption_korean=5*np.random.randn(1000)+53.9
meat_consumption_japan=4*np.random.randn(1000)+32.7

meat_consumption=pd.DataFrame({"한국인":meat_consumption_korean,"일본인":meat_consumption_japan})
```

```python
meat_consumption
```

```python
#한국인 육류소비량 히스토그램
plt.hist(meat_consumption_korean)
plt.xlabel("Korea")
plt.show()
```

```python
#일본인 육류소비량 히스토그램
plt.hist(meat_consumption_japan)
plt.xlabel("Japan")
plt.show()
```

```python
#z-표준화
import scipy.stats as ss

meat_consumption["한국인_표준화"]=ss.zscore(meat_consumption_korean)
meat_consumption["일본인_표준화"]=ss.zscore(meat_consumption_japan)

meat_consumption
```

```python
meat_consumption["한국인_표준화2"]=(meat_consumption_korean-np.mean(meat_consumption_korean))/np.std(meat_consumption_korean)
meat_consumption["일본인_표준화2"]=(meat_consumption_japan-np.mean(meat_consumption_japan))/np.std(meat_consumption_japan)

meat_consumption
```

```python
plt.hist(meat_consumption["한국인_표준화2"])
plt.xlabel("korea")
plt.show()
```

```python
plt.hist(meat_consumption["일본인_표준화2"])
plt.xlabel("japan")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
meat_consumption["한국인_표준화3"]=scaler.fit_transform(meat_consumption[["한국인"]])

meat_consumption["일본인_표준화3"]=scaler.fit_transform(meat_consumption[["일본인"]])

meat_consumption
```

```python
#정규화
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

meat_consumption["한국인_정규화"]=scaler.fit_transform(meat_consumption[["한국인"]])
meat_consumption["일본인_정규화"]=scaler.fit_transform(meat_consumption[["일본인"]])

meat_consumption
```

```python
max_k=np.max(meat_consumption["한국인"])
min_k=np.min(meat_consumption["한국인"])
max_j=np.max(meat_consumption["일본인"])
min_j=np.min(meat_consumption["일본인"])

meat_consumption["한국인_정규화2"]=(meat_consumption["한국인"]-min_k)/(max_k-min_k)

meat_consumption["일본인_정규화2"]=(meat_consumption["일본인"]-min_j)/(max_j-min_j)

meat_consumption
```

```python
import matplotlib.pyplot as plt
data=[["철수",52],["영희",92],["미영",84],["시완",71],["미경",65],["영환",81],["숙경",66],["부영",77],["민섭",73],["보연",74]]
df=pd.DataFrame(data,columns=['이름','수학점수'])

print(np.mean(df["수학점수"]))

plt.hist(df["수학점수"],bins=5,range=[50,100],rwidth=0.9)
plt.show()
df
```

```python
df["등급"]=0

df.loc[(df["수학점수"]<60),"등급"]="F"
df.loc[(df["수학점수"]>=60) & (df["수학점수"]<70),"등급"]="D"
df.loc[(df["수학점수"]>=70) & (df["수학점수"]<80),"등급"]="C"
df.loc[(df["수학점수"]>=80) & (df["수학점수"]<90),"등급"]="B"
df.loc[(df["수학점수"]>90),"등급"]="A"

df
```

```python
df["등급1"]=pd.cut(x=df["수학점수"],bins=[0,60,70,80,90,100],labels=["F",'D','C',"B",'A'],include_lowest=True)

df
```

```python
#qcut:각 범주에 속하는 개수가 동일하도록
df["등급_qcut"]=pd.qcut(x=df["수학점수"],q=5,labels=["F","D",'C','B','A'])
df
```

```python
#결측치 확인
df=pd.read_csv("C://Users//whs38//OneDrive//바탕 화면//archive//titanic.csv")

df.isnull()
```

```python
print(df.isnull().sum())
```

```python
#결측치가 있는 행 모두 제거
df_1=df.dropna(axis=0)
print(df_1.isnull().sum())
print(df_1.shape)
```

```python
#결측치를 평균값으로 대체
print(df["Age"].isnull().sum())
age_mean=df["Age"].mean()
df["Age"].fillna(age_mean,inplace=True)
print(df["Age"].isnull().sum())
```

```python
#결측치를 최빈값으로 대체
from scipy.stats import mode

print(df["Embarked"].isnull().sum())
embarked_mode=df["Embarked"].mode()

df["Embarked"].fillna(embarked_mode[0],inplace=True)
print(df["Embarked"].isnull().sum())
```

```python
#결측치를 인접한 값으로 대체
df["Embarked"].fillna(method="ffill",inplace=True)
#method="ffill" :직전행
#method="bill" : 다음행
```

```python
print(df.groupby("Sex")["Age"].mean())
print(df.groupby("Pclass")["Age"].mean())
```

```python
df.tail()
```

```python
#그룹별 평균값으로 대체
df["Age"].fillna(df.groupby("Pclass")["Age"].transform('mean'),inplace=True)
df.tail()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=10*np.random.randn(200)+50
df=pd.DataFrame({"값":data})
```

```python
df.loc[201]=2
df.loc[202]=100
df.loc[203]=10
df.loc[204]=110

plt.hist(df["값"],bins=20,rwidth=0.9)
plt.show()
```

```python
plt.boxplot(df["값"])
plt.show()
```

```python
Q1=df["값"].quantile(0.25)
Q2=df["값"].quantile(0.5)
Q3=df["값"].quantile(0.75)

IQR=Q3-Q1
print("Q1= " ,Q1)
print("Q2= ",Q2)
print('Q3= ',Q3)
print("IQR= ",IQR)
```

```python
Q1=df["값"].describe()["25%"]
Q2=df["값"].describe()["50%"]
Q3=df["값"].describe()["75%"]

IQR=Q3-Q1

print("Q1= " ,Q1)
print("Q2= ",Q2)
print('Q3= ',Q3)
print("IQR= ",IQR)
```

```python
condition=df["값"]>(Q3+1.5*IQR)
upperoutlier=df[condition]
condition=df["값"]<(Q1-1.5*IQR)
loweroutlier=df[condition]

print(upperoutlier)
print(loweroutlier)
```
